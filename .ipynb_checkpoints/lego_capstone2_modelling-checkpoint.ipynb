```python
target_names = 'set_num set_name theme_id theme_name is_city id part_num color_name part_name part_material is_ninjago'.split()
target_names
```

```python
hyper_table = list()
model = LogisticRegression(class_weight='balanced') # class_weight={0:0.029, 1:0.971})
```

```python
y = df['is_ninjago']
X_train, X_test, y_train, y_test = train_test_split(df[df.columns], y, test_size= 0.1, random_state=1)
```

```python
fn = list()
for i, name in tqdm(enumerate(df.columns), total=len(df.columns)):
    if i > 100:
        break
    if name in target_names:
        continue
    fn.append(name)
    model.fit(X_train[fn], y_train)
    s_train = model.score(X_train[fn], y_train)
    s_test = model.score(X_test[fn], y_test)
    hyper_table.append(dict(desc=str(fn), train_score=s_train, test_score=s_test, C=model.C, model=model, num_features=df[fn].shape[1], num_rows=df[fn].shape[0], last_feature=fn[-1]))
```

```python
ht0 = pd.DataFrame(hyper_table).sort_values('test_score', ascending=False)
ht0
```

```python
print('Precision: %.3f' % precision_score(y_test, y_pred))

print('Recall: %.3f' % recall_score(y_test, y_pred))

print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))

print('F1 Score: %.3f' % f1_score(y_test, y_pred))
```

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.under_sampling import NearMiss
from collections import Counter
from sklearn.metrics import classification_report
```

```python
fn2 = list()
for i, name in enumerate(df.columns):
    if i > 100:
        break
    if name in target_names:
        continue
    fn2.append(name)
    
X = df[fn2]
y = df['is_ninjago']
print(Counter(y))

oversample = RandomOverSampler(sampling_strategy='minority')
X_over, y_over = oversample.fit_resample(X, y)
print(Counter(y_over))

X_train, X_test, y_train, y_test = train_test_split(X_over, y_over, test_size= 0.1, random_state=1)

lr = LogisticRegression() # class_weight={0:0.029, 1:0.971})
lr.fit(X_train, y_train)
s_train = lr.score(X_train, y_train)
s_test = lr.score(X_test, y_test)
y_pred = lr.predict(X_test)
print(s_train)
print(s_test)
print(y_pred)
```

```python
print(Counter(y))

undersample = RandomUnderSampler(sampling_strategy='majority')
X_under, y_under = undersample.fit_resample(X, y)
print(Counter(y_under))

X_train, X_test, y_train, y_test = train_test_split(X_under, y_under, test_size= 0.1, random_state=1)

lr = LogisticRegression() # class_weight={0:0.029, 1:0.971})
lr.fit(X_train, y_train)
s_train = lr.score(X_train, y_train)
s_test = lr.score(X_test, y_test)
y_pred = lr.predict(X_test)
print(s_train)
print(s_test)
print(y_pred)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.1, random_state=1)

print("Before Undersampling, counts of label '1': {}".format(sum(y_train == 1)))
print("Before Undersampling, counts of label '0': {} \n".format(sum(y_train == 0)))

from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 2)

X_train_res, y_train_res = sm.fit_resample(X_train, y_train.ravel())
 
print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))
 
print("After OverSampling, counts of label '1': {}".format(sum(y_train_res == 1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res == 0)))

lr1 = LogisticRegression()
lr1.fit(X_train_res, y_train_res.ravel())
predictions = lr1.predict(X_test)
 
# print classification report
print(classification_report(y_test, predictions))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.1, random_state=1)

print("Before Undersampling, counts of label '1': {}".format(sum(y_train == 1)))
print("Before Undersampling, counts of label '0': {} \n".format(sum(y_train == 0)))
 
# apply near miss
from imblearn.under_sampling import NearMiss
nr = NearMiss()
 
X_train_miss, y_train_miss = nr.fit_resample(X_train, y_train.ravel())
 
print('After Undersampling, the shape of train_X: {}'.format(X_train_miss.shape))
print('After Undersampling, the shape of train_y: {} \n'.format(y_train_miss.shape))
 
print("After Undersampling, counts of label '1': {}".format(sum(y_train_miss == 1)))
print("After Undersampling, counts of label '0': {}".format(sum(y_train_miss == 0)))

# train the model on train set
lr2 = LogisticRegression()
lr2.fit(X_train_miss, y_train_miss.ravel())
predictions = lr2.predict(X_test)
 
# print classification report
print(classification_report(y_test, predictions))
```

```python
is_ninjago_0, is_ninjago_1 = df['is_ninjago'].value_counts()
```

```python
# Separate df
# df_grouped = df.groupby('is_ninjago')
# df_0 = df_grouped.groups[0]
# df_1 = df_grouped.groups[1]

df_0 = df[df['is_ninjago'] == 0]
df_1 = df[df['is_ninjago'] == 1]
print(df_0.shape)
print(df_1.shape)
```

```python
df_under = df_0.sample(is_ninjago_1)

test_under = pd.concat([df_under, df_1], axis=0)

print(test_under['is_ninjago'].value_counts())
```

```python
test_under.head()
```

```python
test_under.shape
```

```python
ht_under = list()
fn_under = list()

model = LogisticRegression() # class_weight='balanced') # class_weight={0:0.029, 1:0.971})

y = test_under['is_ninjago']
X_train, X_test, y_train, y_test = train_test_split(test_under[test_under.columns], y, test_size= 0.1, random_state=1)

df_train = pd.concat([pd.DataFrame(X_train, columns=fn_under), pd.DataFrame(y_train, columns=['is_ninjago'])], axis=1)
df_test = pd.concat([pd.DataFrame(X_test, columns=fn_under), pd.DataFrame(y_test, columns=['is_ninjago'])], axis=1)

for i, name in tqdm(enumerate(test_under.columns), total=len(test_under.columns)):
    if i > 100:
        break
    if name in target_names:
        continue
    fn_under.append(name)
    model.fit(X_train[fn_under], y_train)
    test_scores = []

    for is_ninjago, group in df_test.groupby('is_ninjago'):
        test_scores.append(model.score(X_test[fn_under], y_test))
    
    train_scores = []
    
    for is_ninjago, group in df_train.groupby('is_ninjago'):
        train_scores.append(model.score(X_train[fn_under], y_train))
        
    s_train = model.score(X_train[fn_under], y_train)
    s_test = model.score(X_test[fn_under], y_test)
    ht_under.append(dict(sample='under', 
                         desc=str(fn_under), 
                         train_score=s_train, 
                         test_score=s_test, 
                         test_score_0=test_scores[0], 
                         test_score_1=test_scores[1],
                         train_score_0=train_scores[0],
                         train_score_1=train_scores[1],
                         C=model.C, 
                         model=model, 
                         num_features=test_under[fn_under].shape[1], 
                         num_rows=test_under[fn_under].shape[0], 
                         last_feature=fn_under[-1]))
```

```python
df_ht_under = pd.DataFrame(ht_under).sort_values('test_score', ascending=False)
df_ht_under
```

```python
print(df_ht_under['test_score'].mean())
print(df_ht_under['test_score_0'].mean())
print(df_ht_under['test_score_1'].mean())
print(df_ht_under['train_score'].mean())
print(df_ht_under['train_score_0'].mean())
print(df_ht_under['train_score_1'].mean())
```
